```python
import networkx as nx
import matplotlib.pyplot as plt
import random 
%matplotlib inline 
import itertools
import collections

def undercon(G, k):
    to_be_removed = []
    for n in G:
        if len(G[n]) < k:
            to_be_removed.append(n)

    for node in to_be_removed:       
        G.remove_node(node) # G.remove_node(n)
    
    return G

def subsumed(G, k):
    to_be_removed = []
    all_nodes = set(G)
    for n in G:
        n_edges = set(G[n])
        for m in all_nodes:
            if n == m:
                continue
            m_edges = set(G[m])

            if n_edges.issubset(m_edges):
                to_be_removed.append(n)
                all_nodes.remove(n)
                break

    for node in to_be_removed:
        G.remove_node(node)
    
    return G
           
def create_edge(G, K, node_1, node_2):
    if node_1 != node_2:
        G.add_edge(node_1, node_2)
    
    for clique in nx.cliques_containing_node(G, nodes=node_1):
        if node_2 in clique and len(clique) > K:
            G.remove_edge(node_1, node_2)
            break
        
    return G
    
    
def gen_random_graph_1(N, K):
    # Dit heeft nog steeds N/10 erin dat gewoon verzonnen is.
    # Gaat altijd rond een bepaalde connectivity zitten bij een bepaald aantal nodes.
    G = nx.Graph()
    G.add_nodes_from(range(N))
    
    p = 0.5
    for node_1 in G:
        rnum = random.randrange(N/10)
        for count in range(rnum):
            node_2 = random.randrange(N)
            if random.random() < p:
                G = create_edge(G, K, node_1, node_2)
                
    return G

def gen_random_graph_2(N, K):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    
    for node_1 in G:
        for node_2 in G:
            p = 5 / N
            if random.random() < p:
                G = create_edge(G, K, node_1, node_2)
                
    return G

# Ik wil er eentje die ook verschillende densities veroorzaakt bij eenzelfde aantal nodes.



```

```python
K = 4
G = gen_random_graph_1(200, K)
F = gen_random_graph_2(200, K)


def reduce_graph(G, K):
    while True:
        n_nodes_2 = len(G)
        for reduce in (subsumed, symmetry, undercon):#, symmetry, undercon):
            n_nodes = len(G)
            while True:
                G = reduce(G, K)
                if len(G) == n_nodes:
                    break
                n_nodes = len(G)
        if n_nodes_2 == len(G):
            break
    return G

G = reduce_graph(G, K)
F = reduce_graph(F, K)

nx.draw(G,with_labels = True)
print(nx.number_of_nodes(G))
plt.show()
nx.draw(F,with_labels = True)
print(nx.number_of_nodes(F))
plt.show()
degree = G.degree
```
